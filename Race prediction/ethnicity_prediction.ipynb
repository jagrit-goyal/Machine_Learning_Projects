```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('elon.jpg')
plt.imshow(img[:,:,::-1])
plt.show()
```

```python
result = DeepFace.analyze(img , actions=['race'])
```

```python
print(result)
```

```python
race_dict = result[0]['race']
```

```python
max_race = max(race_dict, key=race_dict.get)
max_value = race_dict[max_race]
```

```python
print(f"Dominant race: {max_race}")
print(f"Max value: {max_value}")
```
