```python
from deepface import DeepFace
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('sad.jpg')
```

```python
plt.imshow(img[:, :, ::-1])
plt.show()
```

```python
result = DeepFace.analyze(img , actions = ['emotion'])
```

```python
print(result)
```

printing the max value

```python
emotion_dict = result[0]['emotion']
```

```python
max_emotion = max(emotion_dict, key=emotion_dict.get)
max_value = emotion_dict[max_emotion]
```

```python
print(f"Dominant emotion: {max_emotion}")
print(f"Max value: {max_value}")
```
